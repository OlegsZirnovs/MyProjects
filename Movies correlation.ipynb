```python
#import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

#Read in the data

df = pd.read_csv(r'C:\Users\MSI\Downloads\movies.csv')
```

```python
#let's look on the data

df.head()
```

```python
# let's see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    
    if (pct_missing > 0.00001):
        df.dropna(inplace=True)    
    print('{} - {}%'.format(col, pct_missing))

num_rows = len(df)
print("Number of rows:", num_rows)
```

```python
df.dtypes
```

```python
#change data type of columns

df['budget'].round().astype('Int64')
df['budget']=df['budget'].astype('Int64')

df['gross'].round().astype('Int64')
df['gross']=df['gross'].astype('Int64')
```

```python
df.head()
```

```python
#Create correct Year column

df['yearcorrect'] = df['released'].astype(str).str.split().str[2]

df
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Drop any duplicates

df.drop_duplicates()
df.head()
```

```python
# Budget high correlation
#company high correlation

```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'],y=df['gross'])
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['gross'] = pd.to_numeric(df['gross'], errors='coerce')
df.dropna(subset=['budget', 'gross'], inplace=True)  # Drop rows with NaN values
df['budget'] = df['budget'].abs()
df['gross'] = df['gross'].abs()

plt.title('Budget vs Gross earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')
plt.show()
```

```python
# Plot budget vs gross using seaborn
df.to_csv('movies.csv', index=False)
data = pd.read_csv(r'C:\Users\MSI\Downloads\movies.csv')
# draw regplot
sns.regplot(x = 'budget', 
            y = 'gross', 
            data = data,
           scatter_kws={"color": "red"}, line_kws={"color":"blue"})
plt.show()
```
